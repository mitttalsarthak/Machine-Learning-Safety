```python
from kprojection import KProjectionCoverage

odd_description = {
    "weather": ["sunny", "rain_fog_snow"],
    "lighting": ["day", "night"],
    "camera": ["normal", "damaged"],
    "scene": ["urban", "unmapped"],
    "speed": ["leq_50", "gt_50"]
}
```

```python
scenarios = [
    {
        "weather": "sunny",
        "lighting": "day",
        "camera": "normal",
        "scene": "urban",
        "speed": "leq_50"
    }
]
```

```python
cov1 = KProjectionCoverage(k=1, desc=odd_description)

cov1.add_scenarios(scenarios)

result1 = cov1.compute()

print(result1)
```

```python
cov2 = KProjectionCoverage(k=2, desc=odd_description)

cov2.add_scenarios(scenarios)

result2 = cov2.compute()

print(result2)
```
